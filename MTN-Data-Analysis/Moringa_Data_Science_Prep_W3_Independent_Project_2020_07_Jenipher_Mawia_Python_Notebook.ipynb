# Exploration and Cleaning

```
# load the dataset and view 
cells_geo_desc = pd.read_excel("cells_geo_description.xlsx")
cells_geo_desc
```

```
# read the dataset and view
cells_geo = pd.read_csv("cells_geo_new.csv")
cells_geo
```

```
# read the excel file and view
df_DESC = pd.read_excel('CDR_description.xlsx')
df_DESC
```

```
# read first telecom dataset and view
df_telcom1 = pd.read_csv('Telcom_dataset.csv')
df_telcom1
#rename column 'PRODUTC' to 'PRODUCT'
df_telcom1.rename(columns = {'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}, inplace = True)
#view the output of changes made
print(df_telcom1.columns)
```

```
# read second telecom dataset and view
df_telcom2 = pd.read_csv('Telcom_dataset2.csv')
df_telcom2
#rename columns
df_telcom2.rename(columns = {'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'CELLID': 'CELL_ID', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'}, inplace = True)
#view the output of changes made
print(df_telcom2.columns)
```

```
# read and view dataset
df_telcom3 = pd.read_csv('Telcom_dataset3.csv')
df_telcom3
#rename columns
df_telcom3.rename(columns = {'SIET_ID': 'SITE_ID', 'CELLID':'CELL_ID'}, inplace = True)
#view the output of changes made
print(df_telcom3.columns)
```

```
#append the three dataframes together and view
new = df_telcom1.append([df_telcom2, df_telcom3])
new
```

```
# drop the first column of the cells_geo
cells_geo= cells_geo.drop(cells_geo.columns[[0]], axis=1)
cells_geo
```

```
# rename the columns
cells_geo.rename(columns = {'SITE_CODE': 'SITE_ID'}, inplace = True)
cells_geo
```

```
# merge all the three datasets onto a new dataframe and view
telcom_geo = pd.merge(new, cells_geo, on='SITE_ID')
telcom_geo
```

**Cleaning**

```
# drop cells that are not necessary
telcom_geo.drop(["CELL_ON_SITE","COUNTRY_A","COUNTRY_B", "CELL_ID_x", "CELL_ID_y"], axis=1)
```

```
#  finding duplicates
telcom_geo.duplicated()

telcom_geo.isnull()

# fill the null values with "."
telcom_geo.fillna(".")
```
